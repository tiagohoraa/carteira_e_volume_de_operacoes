Projeção da carteira de crédito

Aqui vamos considerar a carteira de crédito como somatório de todos os empréstimos realizados pela instituição. Neste sentido, vamos somar o valor bruto e agrupar por mês para modelar a série temporal usando o ARIMA Sazonal.

```python
df=base[['data_operacao','valor_bruto']]
```

```python
df= df.groupby('data_operacao')['valor_bruto'].sum().reset_index()
```

```python
df = df.set_index('data_operacao')
df.head()
```

```python
y=df
```

```python
y = y['valor_bruto'].resample('MS').sum()
```

```python
y.head()
```

```python
df2=pd.DataFrame(y)
df2.head()
```

```python
df2.plot(figsize=(8, 4))
plt.show()
```

```python
decompose = seasonal_decompose(df2, extrapolate_trend=1)

decompose.plot();
```

```python
def Estacionaria(timeseries):
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Estatística do teste','p-value','O critério de informação maximizado ','Número de observações usadas'])
    for key,value in dftest[4].items():
        dfoutput['Valor crítico (%s)'%key] = value
    
    if(dfoutput['Estatística do teste'] < dfoutput['Valor crítico (5%)'] and dfoutput['p-value'] < 0.05):
        print('É estacionária.')
    else:
        print('Não é estacionária.')
```

```python
Estacionaria(df2.valor_bruto)
```

Observando o comportamento sazonal da série e a característica não estacionária, faz-se necessário aplicar o SARIMA para realizar a extrapolação.

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

No próximo passo, realiza-se uma busca para encontrar a melhor configuração dos parâmetros p, d e q.

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal)

            results = mod.fit()

            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

Desta maneira, utiliza-se a configuração que apresentou o menor valor de AIC.

```python
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(0, 1, 1, 12))

results = mod.fit()

print(results.summary().tables[1])
```

Apenas o coeficiente auto regressivo (ar) possui significância, (p-valor < 0.05), e deve entrar no modelo.

```python
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

No gráfico superior direito, a linha laranja do KDE segue de perto a linha N(0,1). Onde N(0,1) é a notação padrão para uma distribuição normal com média 0 e desvio padrão de 1. Esta é uma boa indicação de que os resíduos são normalmente distribuídos.O gráfico qq no canto inferior esquerdo mostra que a distribuição ordenada dos resíduos (pontos azuis) segue a tendência linear das amostras retiradas de uma distribuição normal padrão. Novamente, esta é uma forte indicação de que os resíduos são normalmente distribuídos.
Os resíduos ao longo do tempo (gráfico superior esquerdo) não apresentam nenhuma sazonalidade óbvia e parecem ser ruído branco. Isto é confirmado pelo gráfico de autocorrelação (ou seja, correlograma) no canto inferior direito, que mostra que os resíduos da série temporal têm baixa correlação com versões defasadas de si mesmos. Neste sentido, pode-se concluir que o modelo produz um ajuste satisfatório e que pode ser útil para compreender os dados de série temporal e prever valores futuros.

```python
pred = results.get_prediction(start=pd.to_datetime('2022-01-01'), dynamic=False)
pred_ci = pred.conf_int()

ax = y['2019':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

ax.set_xlabel('data_operacao')
ax.set_ylabel('valor da carteira')
plt.legend()

plt.show()
```

As projeções estão alinhadas com os valores verdadeiros, mostrando comportamento similar.

```python
y_forecasted = pred.predicted_mean
y_truth = y['2022-01-01':]

# mean square error
mse = ((y_forecasted - y_truth) ** 2).mean()
print('O MSE da previsão é {}'.format(round(mse, 2)))
```

```python
#root mean square error

print('O RMSE da previsão é {}'.format(round(np.sqrt(mse), 2)))
```

```python
#projeção do crescimento da carteira de crédito nos próximos 2 anos

pred_uc = results.get_forecast(steps=24)
pred_ci = pred_uc.conf_int()

ax = y.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('data_operacao')
ax.set_ylabel('valor da carteira')

plt.legend()
plt.show()
```

```python
#projeção do valor da carteira de crédito mês a mês nos próximos 2 anos

forecast = pred_uc.predicted_mean

print(forecast)
```
